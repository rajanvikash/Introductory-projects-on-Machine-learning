```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/18_PCA/Exercise/heart.csv")
print(df.shape)
df.head(10)
```

```python
df['RestingECG'].unique()
```

```python
df2  = df[['Sex' , 'ChestPainType' , 'RestingECG','ST_Slope', 'ExerciseAngina']]
df2 = pd.get_dummies(df2 , drop_first=True)
df2
```

```python
df.drop(['Sex' , 'ChestPainType' , 'RestingECG','ST_Slope', 'ExerciseAngina'] , axis = 1, inplace = True)
```

```python

df.head()
```

```python
df = pd.concat([df, df2] , axis = 1 )
df.head()
```

```python
target = df['HeartDisease']
df.drop('HeartDisease' , axis = 1 , inplace = True)
df.head()
```

```python
x = df
y = target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scale = scaler.fit_transform(x)
x_scale
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test  , y_train , y_test = train_test_split(x_scale, y , test_size=0.25  , random_state=45)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
def PrintScore(x_train , x_test , y_train , y_test):
    logreg = LogisticRegression(max_iter=1000 , solver ='liblinear' , multi_class='auto')
    logreg.fit(x_train , y_train)
    log_score = logreg.score(x_test , y_test)
    
    svm = SVC(kernel = 'rbf' , gamma= 'auto' , C =20)
    svm.fit(x_train , y_train)
    svm_score = svm.score(x_test , y_test)
    
    randforest = RandomForestClassifier()
    randforest.fit(x_train , y_train)
    rand_score = randforest.score(x_test , y_test)
    print("LogisiticRegression_score: {} , svm_score: {}  , randomforest_score : {}".format(log_score , svm_score , rand_score))
```

```python
PrintScore(x_train , x_test  , y_train , y_test )
```

```python
from sklearn.decomposition import PCA
pca1 = PCA(0.90)
pca2 = PCA(n_components=10)
x_pca1  = pca1.fit_transform(x_scale)
x_pca2  = pca2.fit_transform(x_scale)
x_pca1.shape , x_pca2.shape
```

```python
PrintScore(x_train , x_test  , y_train , y_test )
```
